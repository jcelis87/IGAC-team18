# In-depth EDA of IGAC databases

## List of cartographic sheets for Chaparral

```python
filenames = os.listdir("INSUMOS_RETO/BASES_TOPONIMIA_CHAPARRAL")
chap_sheets = [x[3:-4] for x in filenames]
```

```python
layers = fiona.listlayers("25k_2017_V2_gdb/25k_2017_V2.gdb")
```

```python
for layer in layers:
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    if len(gdf) > 0:
        print("-"*60)
        print(f"Layer: {layer}, # of rows: {len(gdf)}")
        print(gdf.head(2))
```

```python
for layer in layers:
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    if len(gdf) > 0 and "NOMBRE_GEOGRAFICO" in gdf.columns:
        print(f"Layer: {layer}, # of rows: {len(gdf)}")
```

```python
relevant_layers = []
for layer in layers:
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    if len(gdf) > 1 and "NOMBRE_GEOGRAFICO" in gdf.columns:
        relevant_layers.append(layer)
```

```python
ce = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer="Central_Energia")
ce
```

```python
final_layers = []
for layer in relevant_layers:
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    if len(gdf[gdf.HOJA.isin(chap_sheets)]) > 0 :
        final_layers.append(layer)
        print(layer)
```

```python
def filter_chaparral(gdf):
    return gdf[gdf.HOJA.isin(chap_sheets)]
```

```python
import pandas as pd
import numpy as np

filtered_layers = []
for layer in final_layers:
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    gdf = gdf[gdf.HOJA.isin(chap_sheets)]
    gdf.NOMBRE_GEOGRAFICO.replace("<Null>", np.nan, inplace=True)
    gdf.dropna(subset=["NOMBRE_GEOGRAFICO"], inplace=True)
    if len(gdf) > 0:
        filtered_layers.append(layer)
        print(layer)
```

## Geo Data Frames to work with

```python
def get_gdf(layer):
    gdf = geopandas.read_file("25k_2017_V2_gdb/25k_2017_V2.gdb", driver="FileGDB", layer=layer)
    gdf = gdf[gdf.HOJA.isin(chap_sheets)]
    gdf.NOMBRE_GEOGRAFICO.replace("<Null>", np.nan, inplace=True)
    gdf.dropna(subset=["NOMBRE_GEOGRAFICO"], inplace=True)
    return gdf
```

```python
cp = get_gdf(filtered_layers[0])
ap = get_gdf(filtered_layers[1])
pp = get_gdf(filtered_layers[2])
lg = get_gdf(filtered_layers[3])
ds = get_gdf(filtered_layers[4])
dd = get_gdf(filtered_layers[5])
og = get_gdf(filtered_layers[6])
```

```python
gdfs = dict()
for layer in filtered_layers:
    gdfs[layer] = get_gdf(layer)
```

```python
water_bodies = ["Río","Quebrada"]
def water(geo_name):
    for wb in water_bodies:
        if wb in geo_name:
            return wb
    return np.nan
```
